# Matplotlib 1

```python
import matplotlib.pyplot as plt
```

```python
# %matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x**2
```

```python
list(zip(x, y))
```

```python
# Functional
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("This is the title")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y, "r")

plt.subplot(1, 2, 2)
plt.plot(y, x, "b")
```

```python
# OO
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel("X axis")
axes.set_ylabel("Y axis")
axes.set_title("Title")
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axis1.plot(x, y)
axis2.plot(y, x)
axis1.set_title("My big title")
axis1.set_xlabel("X")
axis1.set_ylabel("Y")
axis2.set_title("My little title")
axis2.set_xlabel("x")
axis2.set_ylabel("y")
```
