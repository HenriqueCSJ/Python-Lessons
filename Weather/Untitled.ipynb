# Weather from Weather Underground

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Weather.csv")
```

```python
df.head()
```

```python
df["DateTime"] = pd.to_datetime(df["DateTime"], infer_datetime_format=True)
```

```python
df.info()
```

```python
df.columns
```

```python
df[['DateTime', 'CurrTemp', 'WSpeedKmh', 'UV', 'Hum']].head()
```

```python
plt.figure(figsize=(20, 3))
sns.lineplot(x="DateTime", y="CurrTemp", data=df)
```

```python
sns.scatterplot(x="Hum", y="CurrTemp", data=df)
```

```python
df["Hour"] = pd.DatetimeIndex(df["DateTime"]).hour
```

```python
df["Hour"] == 12
```

```python
df12PM = df[df["Hour"] == 14]
```

```python
plt.figure(figsize=(20, 3))
sns.lineplot(x="DateTime", y="Hum", data=df12PM)
```
