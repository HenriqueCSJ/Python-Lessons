```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weather = pd.read_csv("weather_pbi.csv")
```

```python
weather.head()
```

```python
weather["Year"] = pd.to_datetime(weather["Year"], format="%Y").dt.year
weather["Month"] = pd.to_datetime(weather["Month"], format="%b").dt.month
weather["Day"] = pd.to_datetime(weather["Day"], format="%d").dt.day
```

```python
sns.pairplot(weather)
```

```python
sns.scatterplot(x="Year", y="Month", data=weather, size="Temp_C")
```

```python
weather.describe()
```

```python
wea = weather.pivot_table(index="Day", columns="Month", values="Temp_C")
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(wea, cmap="viridis", annot=True)
```
