```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("weather_pbi.csv")
```

```python
df["DateTime"] = pd.to_datetime(df["DateTime"], infer_datetime_format=True)
```

```python
df.drop(["TodaysCondition", "HTempC", "LTempC", "CurrCond", "Direction"], axis=1, inplace=True)
```

```python
df.head()
```

```python
sns.pairplot(df, hue="CurrTemp", palette="viridis")
```

```python
sns.lineplot(x="DateTime", y="CurrTemp", data=df)
```

```python
plt.figure(figsize=(12, 7))
sns.lineplot(x="DateTime", y="CurrTemp", data=df)
sns.lineplot(x="DateTime", y="WSpeedKmh", data=df)
sns.lineplot(x="DateTime", y="Hum", data=df)
sns.lineplot(x="DateTime", y="UV", data=df)
```
