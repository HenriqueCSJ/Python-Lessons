# Matplotlib 3

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plot appereance

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="#FF8C00") # Nomes de cores, rgb
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="#FF8C00", linewidth=3, alpha=0.6) # Nomes de cores, rgb
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="#FF8C00", lw=3, ls="steps") # Nomes de cores, rgb
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="#FF8C00", linewidth=3, marker="o", markersize=8)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="#FF8C00", linewidth=3, marker="o", markersize=8,
       markerfacecolor="yellow", markeredgewidth=2, markeredgecolor="black")
```

# Plot range

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="#FF8C00", lw=3, ls="--")
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
